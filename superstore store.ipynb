```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('superstore.xls')
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
# drop unnncesssary columns
df.drop(['Order ID','Product ID','Customer ID','Postal Code','Country'],axis=1,inplace=True)
```

```python
df
```

```python
import matplotlib.pyplot as plt
df.State.unique()
```

```python
results=df.groupby('Category').sum()
category=df.Category.unique()
plt.figure(figsize=(10,8))
plt.xlabel('category',fontsize=18,c='b')
plt.ylabel('profit',fontsize=18,c='b')
plt.bar(category,results['Profit'])
plt.xticks(category,rotation='vertical')
plt.show()
    
    
```

```python
df.groupby('Category').sum().sort_values('Profit',ascending=False)
```

```python
df.groupby('State').sum().sort_values('Profit',ascending=False)
```

```python
# state that had the most profit
profit=df.loc[(df['Profit']>0)]
results=profit.groupby('State').sum()
state2=[State2 for State2 ,df2 in df.groupby('State')]
plt.figure(figsize=(15,6))
plt.xlabel('state',fontsize=18)
plt.ylabel('profit',fontsize=18)
plt.bar(state2,results['Profit'])
plt.xticks(state2,rotation='vertical')
plt.show()
```

```python
df.groupby('Sub-Category').sum().sort_values('Profit',ascending=False)
```

```python
df.groupby(['State','City']).sum().sort_values('Profit',ascending=False).reset_index()
```

```python
#which sub category of technology was highest in sales
tech=df.loc[(df['Category']=='Technology')]
results=tech.groupby('Sub-Category').sum()
#results=[Tech for Tech,tf in tech.groupby('Sub-Category')]

sub=tech['Sub-Category'].unique()
plt.figure(figsize=(15,16))
plt.xlabel(' tech sub category',fontsize=18)
plt.ylabel('sales',fontsize=18)
plt.bar(sub,results['Sales'])
plt.xticks(sub,rotation='vertical')
plt.show()
```

```python
# regions that had the most sales
plt.figure(figsize=(10,8))
region=df.groupby('Region').sum().sort_values('Sales',ascending=False).reset_index()
plt.pie(region.Sales,labels=df.Region.unique(),startangle=90,shadow=True,explode=(0.1,0.1,0.1,0),autopct='%1.1f%%')
plt.title('pie plot')
plt.show()
```

```python
# market share of most profitable region
plt.figure(figsize=(10,8))
region=df.groupby('Region').sum().sort_values('Profit',ascending=False).reset_index()
plt.pie(region.Profit,labels=region.Region.unique(),startangle=90,shadow=True,explode=(0.1,0.1,0.1,0),autopct='%1.1f%%')
plt.title('pie plot')
plt.show()
```

```python
df.Segment.value_counts()
```

```python
x=df.groupby(['Segment']).sum().sort_values('Profit',ascending=False)
x['percentage']=x['Profit']/x.Profit.sum()*100
#consumer profits
plt.pie(x.percentage,labels=df.Segment.unique(),startangle=90,shadow=True,explode=(0.1,0,0),autopct='%1.1f%%')
plt.title(' Segment pie plot',fontsize=18,c='g')
plt.show()
```

```python
# loss making countries
loss=df.loc[(df['Profit']<0)]
#loss.groupby('State').sum().sort_values('Profit',ascending=True)
loss
     
```

```python
loss1=loss.groupby(['Sub-Category']).sum().sort_values('Profit',ascending=True).reset_index()
plt.figure(figsize=(15,6))
plt.title('loss graph')
plt.ylabel('-profit',fontsize=18)
plt.xlabel('sub_category',fontsize=18)
plt.bar(loss1['Sub-Category'].unique(),loss1.Profit)
plt.xticks(loss1['Sub-Category'].unique(),rotation='vertical')
plt.show()
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['year']=df['Order Date'].dt.year
df
```

```python
# most profitable year
pr=df.groupby('year').sum().sort_values('Profit',ascending=False).reset_index()
pr[['year','Profit']]
plt.figure(1,figsize=(15,6))
plt.title('year_profits')
plt.plot(pr['year'],pr['Profit'])
plt.xlabel('year',fontsize=18,c='g')
plt.ylabel('Total Profit',fontsize=18,c='g')
plt.show()
```

```python
# the most popular shipping modes.
per=df['Ship Mode'].value_counts().reset_index()
perc=per['Ship Mode'].sum()
per['percentage']=per['Ship Mode']/perc*100
# plot a graph
plt.figure(figsize=(15,6))
plt.title('shipping modes',c='g')
plt.bar(per['index'],per['percentage'])
plt.ylabel('# percentages',fontsize=16,c='g')
plt.xlabel('shipping modes',fontsize=16,c='g')
plt.show()


```
